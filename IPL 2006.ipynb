**BATSMAN** **ANALYSIS**


```python
e_batsman=input('Enter Batsman Name: ')
def batsman_analysis(e_batsman):
    count_4=0
    count_6=0
    for i,name in enumerate(df['batsman']):
        if name==e_batsman:
            if df['batsman_runs'][i]==4:
                count_4+=1
            if df['batsman_runs'][i]==6:
                count_6+=1
    return count_4,count_6
fours,sixes=batsman_analysis(e_batsman)
print(f'{e_batsman} had hit {fours} fours and {sixes} sixes.')
```

**BOWLER** **ANALYSIS**


```python
e_bowler=input('Enter Bowler Name: ')
def bowler_analysis(e_bowler):
    ext_r=0
    for i,name in enumerate(df['bowler']):
        if name==e_bowler:
            if df['extra_runs'][i]==1:
                ext_r+=1
            elif df['extra_runs'][i]==2:
                ext_r+=2
            elif df['extra_runs'][i]==3:
                ext_r+=3
            elif df['extra_runs'][i]==4:
                ext_r+=4
            elif df['extra_runs'][i]==5:
                ext_r+=5
            elif df['extra_runs'][i]==7:
                ext_r+=7
    return ext_r
extra=bowler_analysis(e_bowler)
print(f'{e_bowler} had given {extra} extra runs 😢.')
```

**TEAM** **ANALYSIS**

```python
e_team=input('Enter Team Name: ')
def team_analysis(e_team):
    count_4=0
    count_6=0
    for i,name in enumerate(df['batting_team']):
        if name==e_team:
            if df['batsman_runs'][i]==4:
                count_4+=1
            if df['batsman_runs'][i]==6:
                count_6+=1
    return count_4,count_6
fours,sixes=team_analysis(e_team)
print(f'{e_team} had hit {fours} fours and {sixes} sixes.'if e_team != 'RCB' else f'{e_team} had hit {fours} fours and {sixes} sixes🗿🗿🗿.' )
```

```python
t_year=int(input('Enter Year: '))
t_team1=input('Enter Team 1: ')
t_team2=input('Enter Team 2: ')
def team_match(t_year,t_team1,t_team2):
    flag=0
    for i in range(636):
        if dfm['season'][i]==t_year and dfm['team1'][i]==t_team1 and dfm['team2'][i]==t_team2:
            flag=1
            print(f'\nMatch: {dfm["team1"][i]} vs {dfm["team2"][i]} \nVenue: {dfm["venue"][i]}, {dfm["city"][i]}. \n{dfm["toss_winner"][i]} decided to {dfm["toss_decision"][i]}. \n{dfm["winner"][i]} won by {dfm["win_by_runs"][i]} runs. \nPlayer of the match: {dfm["player_of_match"][i]}' if dfm['win_by_runs'][i] != 0 else f'\nMatch: {dfm["team1"][i]} vs {dfm["team2"][i]} \nVenue: {dfm["venue"][i]}, {dfm["city"][i]}. \n{dfm["toss_winner"][i]} decided to {dfm["toss_decision"][i]}. \n{dfm["winner"][i]} won by {dfm["win_by_wickets"][i]} wickets. \nPlayer of the match: {dfm["player_of_match"][i]}'  )
            break
    if flag==0:
        print('Match Not Found')

team_match(t_year,t_team1,t_team2)


```
